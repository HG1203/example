```python
import matplotlib.pyplot as plt
import pandas as pd
import math
import numpy as np
import os
```

```python
filePath = r"E:/断裂功/"
filenames = os.listdir(filePath)

for filename in filenames:
    path = filePath+filename
    print(path)
    data = pd.read_excel(path)
    n=int(len(data.columns)/8)
    
    for i in range (0,n):
        m = data.iloc[:,0+i*8][9:]
        m = m.fillna(0).max()
        x = data.iloc[:,2+i*8][9:m]
        y = data.iloc[:,1+i*8][9:m]
        plt.plot(x,y)
        plt.show()

        h = data.iloc[2,1+i*8]
        b = data.iloc[2,3+i*8]


        S=np.trapz(y,x)#积分面积

        Wf = int((S/(2*b*(h-3.25)))*1000)
        print('断裂功为：'+str(Wf))


        data.iloc[5,0+i*8] = '断裂功'
        data.iloc[5,1+i*8] = Wf
        
        data.to_excel(r''+path,index=None)
```
